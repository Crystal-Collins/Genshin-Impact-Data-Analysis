```python
Beginner20 = Level_20.loc[[1,15,113,155,267,505]]
Beginner20
#I could drop the rows but instead I call the lowest base atk of level 20 for each character.
```

```python
#Gen_melt = pd.melt(Gen_drop, id_vars="Lv", var_name="Character", value_name="Base ATK")
#Gen_melt
#sns.catplot(x='Lv', y='Character', hue=("husl", 8), data=Gen_drop, kind='bar')
sns.barplot(x = 'Character', y = 'Base ATK', hue = 'Lv', data = Gen_drop);
#I notice there is a difference in increased Base ATK between levels, depending on the character.
```

```python
Gacha_20 = Level_20.loc[[57,85,169,197,225,239]]
Gacha_20
#Next I select a few additional level 20 characters, in the spirit of gacha
```

```python
Char_tot = pd.concat([Gacha_20, Beginner20])
Char_tot.head(12) 
#I concat the two dataframes to next visualize
```

```python
sns.relplot(data=Char_tot, x="Base ATK", y="Base HP", col="Lv", hue="Character",
            kind="scatter", style = "Character", s = 100, palette = ("icefire"));
#Between all the characters, we can now easily visualize which characters have 
#the strongest attack and highest health of level 20.
```
